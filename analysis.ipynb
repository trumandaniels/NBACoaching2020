## Creating an all-in-one metric measuring the quality of shots based on the shotclock

I'm going to start by finding the average league distribution for 2pt and 3pt shots, seperated by shotclock bin:

```python
league2pt2422 = shot2422_19["2FGA"].sum()
league2pt2218 = shot2218_19["2FGA"].sum()
league2pt1815 = shot1815_19["2FGA"].sum()
league2pt157 = shot157_19["2FGA"].sum()
league2pt74 = shot74_19["2FGA"].sum()
league2pt40 = shot40_19["2FGA"].sum()

total_2pts = league2pt2422+league2pt2218+league2pt1815+league2pt157+league2pt74+league2pt40

league2pt2422_ratio = league2pt2422/total_2pts
league2pt2218_ratio = league2pt2218/total_2pts
league2pt1815_ratio = league2pt1815/total_2pts
league2pt157_ratio = league2pt157/total_2pts
league2pt74_ratio = league2pt74/total_2pts
league2pt40_ratio = league2pt40/total_2pts
```

```python
league3pt2422 = shot2422_19["3PA"].sum()
league3pt2218 = shot2218_19["3PA"].sum()
league3pt1815 = shot1815_19["3PA"].sum()
league3pt157 = shot157_19["3PA"].sum()
league3pt74 = shot74_19["3PA"].sum()
league3pt40 = shot40_19["3PA"].sum()

total_3pts = league3pt2422+league3pt2218+league3pt1815+league3pt157+league3pt74+league3pt40

league3pt2422_ratio = league3pt2422/total_3pts
league3pt2218_ratio = league3pt2218/total_3pts
league3pt1815_ratio = league3pt1815/total_3pts
league3pt157_ratio = league3pt157/total_3pts
league3pt74_ratio = league3pt74/total_3pts
league3pt40_ratio = league3pt40/total_3pts
```

```python
allshot_2pt_ratio = total_2pts/(total_3pts + total_2pts)
allshot_3pt_ratio = total_3pts/(total_3pts + total_2pts)
```

```python
df_shotclock_avg_dist = pd.DataFrame({"2pt": [100*allshot_2pt_ratio,
                                     100*league2pt2422_ratio, 
                                     100*league2pt2218_ratio,
                                     100*league2pt1815_ratio,
                                     100*league2pt157_ratio,
                                     100*league2pt74_ratio,
                                     100*league2pt40_ratio,
                                    ],
             "3pt": [100*allshot_3pt_ratio,
                                     100*league3pt2422_ratio, 
                                     100*league3pt2218_ratio,
                                     100*league3pt1815_ratio,
                                     100*league3pt157_ratio,
                                     100*league3pt74_ratio,
                                     100*league3pt40_ratio,
                                    ],
             "Time Left on Shot Clock": ["All Shots",
                                         "24-22 Seconds",
                                         "22-18 Seconds",
                                         "18-15 Seconds",
                                         "15-7 Seconds",
                                         "7-4 Seconds",
                                         "4-0 Seconds",
                                        ]},
             )



df_shotclock_avg_dist
```

```python
shot_melted = pd.melt(df_shotclock_avg_dist, id_vars="Time Left on Shot Clock",\
                           var_name="Shot Type", value_name="Shot %")

plt.subplots(figsize=(30,20))

g = sns.barplot(x="Time Left on Shot Clock", y="Shot %", hue="Shot Type", palette="Blues_d", data=shot_melted)
g.tick_params(axis='both', which='major', pad=10, size=15)

g.set_xlabel('Seconds Left on the Shot Clock',fontsize=40)
g.set_ylabel('Percentage of Shots %',fontsize=40)

g.set_yticklabels([0,10, 20, 30, 40, 50, 60], size=30)
g.set_xticklabels(g.get_xticklabels(), rotation=0, ha="center", fontsize=30)


g.set_title("When Teams Shoot: Average Offensive Shot Clock Distribution", fontsize=55, pad=20)
g.text(3, 58, "code: github.com/trumandaniels/NBACoaching2020\ntwitter: @trumanadaniels\ndata source: stats.nba.com\ndata is from the '19-'20 season", fontsize=25)

plt.legend(fontsize=30)
```

Using the same process on all the data, we could find the shot clock distributions of all teams (you get the idea).

```python
off_shotclock_df = pd.DataFrame(columns=["2pt shots total", "3pt shots total",
                      "% of 2pt shots 24-22", "% of 2pt shots 22-18",
                      "% of 2pt shots 18-15", "% of 2pt shots 15-7",
                      "% of 2pt shots 7-4", "% of 2pt shots 4-0",
                      "% of 3pt shots 24-22", "% of 3pt shots 22-18",
                      "% of 3pt shots 18-15", "% of 3pt shots 15-7",
                      "% of 3pt shots 7-4", "% of 3pt shots 4-0"],
             index=shot2422_19.index)

off_shotclock_df["2pt shots total"]=shot2422_19["2FGA"]+shot2218_19["2FGA"]+shot1815_19["2FGA"]+shot157_19["2FGA"]+shot74_19["2FGA"]+shot40_19["2FGA"]
off_shotclock_df["3pt shots total"]=shot2422_19["3PA"]+shot2218_19["3PA"]+shot1815_19["3PA"]+shot157_19["3PA"]+shot74_19["3PA"]+shot40_19["3PA"]
off_shotclock_df["% of 2pt shots 24-22"] = 100*shot2422_19["2FGA"]/off_shotclock_df["2pt shots total"]
off_shotclock_df["% of 2pt shots 22-18"] = 100*shot2218_19["2FGA"]/off_shotclock_df["2pt shots total"]
off_shotclock_df["% of 2pt shots 18-15"] = 100*shot1815_19["2FGA"]/off_shotclock_df["2pt shots total"]
off_shotclock_df["% of 2pt shots 15-7"] = 100*shot157_19["2FGA"]/off_shotclock_df["2pt shots total"]
off_shotclock_df["% of 2pt shots 7-4"] = 100*shot74_19["2FGA"]/off_shotclock_df["2pt shots total"]
off_shotclock_df["% of 2pt shots 4-0"] = 100*shot40_19["2FGA"]/off_shotclock_df["2pt shots total"]

off_shotclock_df["% of 3pt shots 24-22"] = 100*shot2422_19["3PA"]/off_shotclock_df["3pt shots total"]
off_shotclock_df["% of 3pt shots 22-18"] = 100*shot2218_19["3PA"]/off_shotclock_df["3pt shots total"]
off_shotclock_df["% of 3pt shots 18-15"] = 100*shot1815_19["3PA"]/off_shotclock_df["3pt shots total"]
off_shotclock_df["% of 3pt shots 15-7"] = 100*shot157_19["3PA"]/off_shotclock_df["3pt shots total"]
off_shotclock_df["% of 3pt shots 7-4"] = 100*shot74_19["3PA"]/off_shotclock_df["3pt shots total"]
off_shotclock_df["% of 3pt shots 4-0"] = 100*shot40_19["3PA"]/off_shotclock_df["3pt shots total"]

#off_shotclock_df["Points Scored Per 100 poss With League Average Shooting"]

off_shotclock_df["2pt 24-22 Expected Points w/Avg Players"] = off_shotclock_df["% of 2pt shots 24-22"]*shot2pt2422_expectedvalue
off_shotclock_df["2pt 22-18 Expected Points w/Avg Players"] = off_shotclock_df["% of 2pt shots 22-18"]*shot2pt2218_expectedvalue
off_shotclock_df["2pt 18-15 Expected Points w/Avg Players"] = off_shotclock_df["% of 2pt shots 18-15"]*shot2pt1815_expectedvalue
off_shotclock_df["2pt 15-7 Expected Points w/Avg Players"] = off_shotclock_df["% of 2pt shots 15-7"]*shot2pt157_expectedvalue
off_shotclock_df["2pt 7-4 Expected Points w/Avg Players"] = off_shotclock_df["% of 2pt shots 7-4"]*shot2pt74_expectedvalue
off_shotclock_df["2pt 4-0 Expected Points w/Avg Players"] = off_shotclock_df["% of 2pt shots 4-0"]*shot2pt40_expectedvalue

off_shotclock_df["3pt 24-22 Expected Points w/Avg Players"] = off_shotclock_df["% of 3pt shots 24-22"]*shot3pt2422_expectedvalue
off_shotclock_df["3pt 22-18 Expected Points w/Avg Players"] = off_shotclock_df["% of 3pt shots 22-18"]*shot3pt2218_expectedvalue
off_shotclock_df["3pt 18-15 Expected Points w/Avg Players"] = off_shotclock_df["% of 3pt shots 18-15"]*shot3pt1815_expectedvalue
off_shotclock_df["3pt 15-7 Expected Points w/Avg Players"] = off_shotclock_df["% of 3pt shots 15-7"]*shot3pt157_expectedvalue
off_shotclock_df["3pt 7-4 Expected Points w/Avg Players"] = off_shotclock_df["% of 3pt shots 7-4"]*shot3pt74_expectedvalue
off_shotclock_df["3pt 4-0 Expected Points w/Avg Players"] = off_shotclock_df["% of 3pt shots 4-0"]*shot3pt40_expectedvalue

off_shotclock_df["2pt Expected Points w/Avg Players"] = off_shotclock_df["2pt 24-22 Expected Points w/Avg Players"]+off_shotclock_df["2pt 22-18 Expected Points w/Avg Players"]+off_shotclock_df["2pt 18-15 Expected Points w/Avg Players"]+off_shotclock_df["2pt 15-7 Expected Points w/Avg Players"]+off_shotclock_df["2pt 7-4 Expected Points w/Avg Players"]+off_shotclock_df["2pt 4-0 Expected Points w/Avg Players"]
off_shotclock_df["3pt Expected Points w/Avg Players"] = off_shotclock_df["3pt 24-22 Expected Points w/Avg Players"]+off_shotclock_df["3pt 22-18 Expected Points w/Avg Players"]+off_shotclock_df["3pt 18-15 Expected Points w/Avg Players"]+off_shotclock_df["3pt 15-7 Expected Points w/Avg Players"]+off_shotclock_df["3pt 7-4 Expected Points w/Avg Players"]+off_shotclock_df["3pt 4-0 Expected Points w/Avg Players"]
off_shotclock_df["Offensive Shotclock Metric"] = off_shotclock_df["2pt Expected Points w/Avg Players"]*(off_shotclock_df["2pt shots total"]/(off_shotclock_df["2pt shots total"]+off_shotclock_df["3pt shots total"])) + off_shotclock_df["3pt Expected Points w/Avg Players"]*(off_shotclock_df["3pt shots total"]/(off_shotclock_df["2pt shots total"]+off_shotclock_df["3pt shots total"]))
```

```python
off_shotclock_df = off_shotclock_df.sort_values("Offensive Shotclock Metric")
off_shotclock_df["Team"] = off_shotclock_df.index
```
